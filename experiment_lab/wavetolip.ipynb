```python
import os
import requests
from tqdm import tqdm

# Step 1: Create the checkpoints folder
os.makedirs("checkpoints", exist_ok=True)

# Step 2: URL of the pretrained model
url = "https://huggingface.co/spaces/anzorq/wav2lip-gan/resolve/main/wav2lip_gan.pth"
file_path = "checkpoints/wav2lip_gan.pth"

# Step 3: Stream download with progress bar
response = requests.get(url, stream=True)
total_size = int(response.headers.get('content-length', 0))

with open(file_path, "wb") as f, tqdm(
    desc="Downloading wav2lip_gan.pth",
    total=total_size,
    unit='B',
    unit_scale=True,
    unit_divisor=1024,
) as bar:
    for data in response.iter_content(chunk_size=1024):
        f.write(data)
        bar.update(len(data))

print("✅ Model downloaded to:", file_path)
```

```python
assert os.path.exists("checkpoints/wav2lip_gan.pth"), "Missing model file"
assert os.path.exists("output/person.jpg"), "Missing face image"
assert os.path.exists("output/output.mp3"), "Missing audio"
```
